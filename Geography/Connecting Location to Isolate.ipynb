```python
import pandas as pd
import numpy as np
import json
import pycountry
import pycountry_convert as pc
```

```python
ip_to_location = pd.read_csv('/home/kin672/gentb-summer22/Geography/ip_to_location_EXTENDED.csv')
uuid_to_ip = pd.read_csv('/home/kin672/gentb-summer22/Geography/uuid_to_ip.csv')
uuid_to_folder_temp = pd.read_csv('/home/kin672/gentb-summer22/Geography/uuid_to_folder.csv', header = None)
with open('/home/kin672/gentb-summer22/Isolating Relevant Files/7-18 Output/relevant_file_names_7-18.json') as f:
    folder_to_isolate = json.load(f)
```

```python
# Properly transpose the folder-to-isolate converter so that there is one column of isolates and another of folders
folder_to_isolate = pd.DataFrame(folder_to_isolate).transpose().reset_index()[['index', 'Folder']]
folder_to_isolate.columns = ['Isolate', 'Folder']
```

```python
# Pull only the unique uuids
valid_uuids = uuid_to_ip[uuid_to_ip['uuid'] != 'No UUID'].drop_duplicates().reset_index(drop = True)
```

```python
# Drop any uuids that don't have an associated ip address with valid location
valid_uuids = valid_uuids[[ip in ip_to_location.ip.unique() for ip in valid_uuids.ip]].reset_index(drop = True)
```

```python
valid_uuids['location'] = [ip_to_location.loc[ip_to_location['ip'] == ip, 'location'].values[0] for ip in valid_uuids.ip]
valid_uuids['City'] = [ip_to_location.loc[ip_to_location['ip'] == ip, 'City'].values[0] for ip in valid_uuids.ip]
valid_uuids['State/Region'] = [ip_to_location.loc[ip_to_location['ip'] == ip, 'State/Region'].values[0] for ip in valid_uuids.ip]
valid_uuids['Country'] = [ip_to_location.loc[ip_to_location['ip'] == ip, 'full_country'].values[0] for ip in valid_uuids.ip]
valid_uuids['Continent'] = [ip_to_location.loc[ip_to_location['ip'] == ip, 'Continent'].values[0] for ip in valid_uuids.ip]
```

```python
uuid_to_folder_temp['folder'] = uuid_to_folder_temp.iloc[:, 1].str.split('/', expand = True)[5]
uuid_to_folder_temp.columns = ['index', 'full_folder', 'uuid', 'folder']
uuid_to_folder = uuid_to_folder_temp[['uuid', 'folder']]
```

```python
isolate_to_uuid = folder_to_isolate[['Isolate', 'Folder']]
isolate_to_uuid['uuid'] = [uuid_to_folder.loc[uuid_to_folder['folder'] == folder,'uuid'].values[0] for folder in isolate_to_uuid['Folder']]
```

```python
# Only include valid UUIDs
isolate_to_uuid = isolate_to_uuid[[uuid in valid_uuids.uuid.values for uuid in isolate_to_uuid.uuid.values]].reset_index(drop = True)
# Add location info
isolate_to_uuid['ip'] = [valid_uuids.loc[valid_uuids['uuid'] == uuid, 'ip'].values[0] for uuid in isolate_to_uuid.uuid]
isolate_to_uuid['Location'] = [valid_uuids.loc[valid_uuids['uuid'] == uuid, 'location'].values[0] for uuid in isolate_to_uuid.uuid]
```

```python
isolate_to_location = isolate_to_uuid[['Isolate', 'Location']]
breakdown = isolate_to_location.Location.str.split(', ', expand = True)
breakdown.columns = ['City', 'State/Region', 'Abbrev_Country']
breakdown['Country'] = [pycountry.countries.get(alpha_2=str(country)).name for country in breakdown['Abbrev_Country']]
```

```python
def country_to_continent(country_name):
    country_alpha2 = pc.country_name_to_country_alpha2(country_name)
    country_continent_code = pc.country_alpha2_to_continent_code(country_alpha2)
    country_continent_name = pc.convert_continent_code_to_continent_name(country_continent_code)
    return country_continent_name

breakdown['Continent'] = [country_to_continent(country) for country in breakdown['Country']]

location_to_lat_long = ip_to_location[['location', 'lat', 'lng']].drop_duplicates()
isolate_to_location['Latitude'] = [location_to_lat_long.loc[location_to_lat_long['location'] == location, 'lat'].values[0] for location in isolate_to_location.Location]
isolate_to_location['Longitude'] = [location_to_lat_long.loc[location_to_lat_long['location'] == location, 'lng'].values[0] for location in isolate_to_location.Location]
```

```python
isolate_to_location = pd.concat([isolate_to_location[['Isolate', 'Location', 'Latitude', 'Longitude']], breakdown[['City', 'State/Region', 'Country', 'Continent']]], axis = 1)
```

```python
isolate_to_location.to_csv('/home/kin672/gentb-summer22/Geography/ISOLATE_TO_LOCATION.csv', index = False)
```
