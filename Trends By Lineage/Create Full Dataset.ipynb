```python
import pandas as pd
import numpy as np
import json
import glob, os, subprocess
import matplotlib.pyplot as plt

summarized_df = pd.read_csv('/home/kin672/gentb-summer22/Creating Summaries from Relevant Files/7-18 Output/summarized_full_7-18.csv', usecols = [1, 2, 3, 4, 5])
```

```python
summarized_df['Total Isolates'] = summarized_df['Number of Resistant Strains'] + summarized_df['Number of Susceptible Strains']
```

```python
summarized_df
```

```python
# All of the individual lineages
all_lineages = summarized_df.loc[:, 'Lineage'].str.split(',', expand=True)
all_lineages = all_lineages.rename(columns = {0: 'Lineage 1', 1: 'Lineage 2', 2: 'Lineage 3', 3:'Lineage 4', 4:'Lineage 5', 5:'Lineage 6', 6:'Lineage 7', 7:'Lineage 8', 8:'Lineage 9'})
all_lineages
```

```python
# Get the number of lineages
summarized_df['Number_of_Lineages'] = pd.Series()
for index, row in all_lineages.iterrows():
    summarized_df.at[index, 'Number_of_Lineages'] = len(row.unique()) - 1
summarized_df
```

```python
# Get the primary lineage family
summarized_df['Primary Lineage Family'] = [str(long)[0:1] if ',' not in str(long) else 'None' if pd.isna(long) else 'Multiple' for long in summarized_df.Lineage]
summarized_df.loc[summarized_df['Primary Lineage Family'] == 'B', 'Primary Lineage Family'] = 'Bovine'
summarized_df.loc[summarized_df['Primary Lineage Family'] == 'n', 'Primary Lineage Family'] = 'None'

```

```python
summarized_df
```

```python
summarized_df.to_csv('/home/kin672/gentb-summer22/full_submission_data_7-24.csv', index = False)
```

```python
# Simplify further
simplified_df = summarized_df.loc[:, ['Drug', 'Primary Lineage Family', 'Number of Resistant Strains', 'Number of Susceptible Strains','Total Isolates']].groupby(['Drug', 'Primary Lineage Family']).sum()
```

```python
simplified_df['Proportion Resistant'] = simplified_df['Number of Resistant Strains']/simplified_df['Total Isolates']
```

```python
simplified_df = simplified_df.reset_index()
simplified_df.to_csv('/home/kin672/gentb-summer22/submission_data_by_primary_lin_7-24.csv', index = False)
```
